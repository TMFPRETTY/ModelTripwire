# ModelTripwire Baseline Analysis

Example notebook for inspecting exported baseline results.

```python
from pathlib import Path
import json
import pandas as pd

project_root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
results_path = project_root / 'outputs' / 'latest' / 'results.json'
rows = json.loads(results_path.read_text())
df = pd.DataFrame(rows)
df[['prompt_id', 'category', 'refusal_score', 'compliance_score', 'tripwire_count']]
```

```python
df.groupby('category')[['refusal_score', 'compliance_score', 'tripwire_count']].mean().sort_values('tripwire_count', ascending=False)
```
